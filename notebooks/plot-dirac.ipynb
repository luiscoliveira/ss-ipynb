Desenhar gráficos com Diracs em tempo contínuo

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

# Move left y-axis and bottom x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.xlim(-2,2)
plt.ylim(-2,2)

dirac = (0, 1)

plt.axhline(y=0, color='blue')
ax.annotate("", xy=(dirac[0], dirac[1]), xytext=(dirac[0], 0), arrowprops=dict(arrowstyle="->", color='blue'))
ax.annotate(str(dirac[1]), xy=(0, 0), xytext=(dirac[0], dirac[1]))

plt.savefig('plot-dirac.png')
```
